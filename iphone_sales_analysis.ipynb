```python
highest_rated['Product Name'].value_counts()
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated['Number Of Ratings'].sort_index()
figure = px.bar(highest_rated, x=label, y=counts, title= "number of rating of highest rated iphone")
figure.show()
```

```python
#which iphone has the highest number of reviews on flipkart?
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated['Number Of Reviews'].sort_index()
figure = px.bar(highest_rated, x=label, y=counts, title= "Number Of Reviews of highest rated iphone")
figure.show()
```
